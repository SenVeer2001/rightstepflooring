# Workiz Application - Screenshot Analysis

This notebook analyzes all screenshots from the sampleimages folder to document the UI/UX design, components, and specifications for the Workiz application.

```python
import os
import sys
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from datetime import datetime
import json

# Set up paths
images_dir = Path("/Users/shakti/Documents/caasaaproject/workiz/sampleimages")
print(f"Screenshots directory: {images_dir}")
print(f"Directory exists: {images_dir.exists()}")

# Get all PNG files sorted by modification time
image_files = sorted(images_dir.glob("*.png"), key=lambda x: x.stat().st_mtime)
print(f"\nTotal screenshots found: {len(image_files)}")
for i, img_file in enumerate(image_files, 1):
    print(f"{i}. {img_file.name}")
```

## Load and Display Screenshots

Let's display each screenshot to analyze the layout and UI elements

```python
# Display each screenshot
fig = plt.figure(figsize=(20, 100))
for idx, img_file in enumerate(image_files, 1):
    ax = plt.subplot(10, 3, idx)
    img = Image.open(img_file)
    ax.imshow(img)
    ax.set_title(f"{idx}. {img_file.name}\n{img.size[0]}x{img.size[1]}px", fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.savefig("/Users/shakti/Documents/caasaaproject/workiz/screenshots_overview.png", dpi=100, bbox_inches='tight')
plt.show()
print("Screenshot overview generated!")
```

## Detailed Screenshot Analysis

Now analyzing each screenshot individually with detailed observations about layout, UI elements, colors, and structure.

```python
# Create detailed analysis for each screenshot
analysis_data = {}

# For each screenshot, we'll analyze and document
for idx, img_file in enumerate(image_files, 1):
    img = Image.open(img_file)
    width, height = img.size
    
    # Get pixel data for color analysis
    pixels = img.getdata()
    
    analysis_data[idx] = {
        "filename": img_file.name,
        "path": str(img_file),
        "dimensions": f"{width}x{height}",
        "size_mb": img_file.stat().st_size / (1024 * 1024),
        "mode": img.mode,
    }

# Display the analysis data
import pandas as pd
df = pd.DataFrame(analysis_data).T
print("Screenshot Metadata:")
print(df.to_string())
print(f"\nTotal Screenshots: {len(analysis_data)}")
```
